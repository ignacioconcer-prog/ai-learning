## 2. Your first data — NumPy arrays

NumPy arrays are the building block of all ML. Think of them as supercharged lists.

````python
# A simple array of numbers
scores = np.array([85, 92, 78, 90, 88, 76, 95, 83])

print(f'Scores:  {scores}')
print(f'Average: {scores.mean():.1f}')
print(f'Highest: {scores.max()}')
print(f'Lowest:  {scores.min()}')
````

## 3. Working with data — Pandas DataFrames

Pandas DataFrames are like Excel spreadsheets in Python.

````python
# Create a small dataset (like a product backlog)
data = {
    'feature':    ['Login page', 'Dashboard', 'API integration', 'Mobile app', 'Analytics'],
    'priority':   [1, 2, 3, 4, 5],
    'effort_days':[3, 8, 12, 20, 15],
    'value_score':[90, 85, 70, 95, 60]
}

df = pd.DataFrame(data)
df['roi'] = df['value_score'] / df['effort_days']  # Simple ROI calculation
df_sorted = df.sort_values('roi', ascending=False)

print('Features ranked by ROI (value per effort day):')
df_sorted
````

## 4. Your first chart — Matplotlib

Visualizations help communicate data — critical for AI/ML work.

````python
fig, ax = plt.subplots(figsize=(8, 4))
bars = ax.barh(df['feature'], df['value_score'], color='steelblue')
ax.set_xlabel('Value Score')
ax.set_title('Feature Value Scores')
ax.bar_label(bars, padding=3)
plt.tight_layout()
plt.show()
````

## 5. Calling Claude (Anthropic API)

Before running this cell, create a `.env` file in this folder with: hola caracola
```
ANTHROPIC_API_KEY=your_key_here
```
Get a key at: https://console.anthropic.com

````python
import os
from dotenv import load_dotenv
import anthropic
import sys

load_dotenv()

api_key = os.getenv('ANTHROPIC_API_KEY')
if not api_key:
    print('Set ANTHROPIC_API_KEY in a .env file first.')
else:
    client = anthropic.Anthropic(api_key=api_key)
    message = client.messages.create(
        model='claude-sonnet-4-6',
        max_tokens=256,
        messages=[{'role': 'user', 'content': 'In one sentence, what is machine learning?'}]
    )
    print('Claude says:', message.content[0].text)
````
